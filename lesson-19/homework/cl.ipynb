```python
import pandas as pd
import sqlite3
with sqlite3.connect('data/chinook.db') as con:
    df_cus=pd.read_sql("Select * from customers", con=con
    )
with sqlite3.connect('data/chinook.db') as con:
    df_inv=pd.read_sql("Select * from invoices", con=con
    )
df_new_sql=pd.merge(df_inv, df_cus, how='inner', on='CustomerId')
df_sql_grouped=df_new_sql.groupby('CustomerId')
df_sql_grouped.count()
```

```python
def isChild(Age):
    if Age>18:
        return 'Adult'
    else:
        return'Child'
df_tit['Age_Group']=df_tit['Age'].apply(isChild)
```

```python
def norm(salary):
    return (salary-np.min(salary))/(np.max(salary)-np.min(salary))
```

```python
def mov_func(duration):
    if duration<60:
        return 'short'
    elif duration<120:
        return 'medium'
    else:
        return 'long'
print(df_mov['duration'].apply(mov_func))
```

```python
def tit_pipe(df):
    df.loc[df['Survived']==1]
    df.Age.fillna(df.Age.mean)
    df['Fare_Per_Age']=df.Fare/df.Age
df_tit.pipe(tit_pipe)
```

```python
def pipe_flight(df):
    df['DepDelay']=pd.to_numeric(df['DepDelay'])
    df.loc[df['DepDelay']>30].shape
    df['AirTime']=pd.to_numeric(df['AirTime'])
    df['Delay_Per_Hour']=df['DepDelay']/df["AirTime"]
```

```python
piped_df_flight=df_flight.pipe(pipe_flight)
```
