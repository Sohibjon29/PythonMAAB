```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
x=np.arange(-10, 10, 0.01)
y=x**2-4*x+4
plt.plot(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
x=np.arange(-10, 2*np.pi, 0.01)
y=np.sin(x)
z=np.cos(x)
plt.figure(figsize=(8, 1))
plt.plot(x, y)
plt.plot(x,z)
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.barh(categories, values, color='orange', height=0.3)
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```
