```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
x=np.arange(-10, 10, 0.01)
y=x**2-4*x+4
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
x=np.arange(-10, 2*np.pi, 0.01)
y=np.sin(x)
z=np.cos(x)
plt.figure(figsize=(8, 1))
plt.plot(x, y, label='sin(x)')
plt.plot(x,z, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x=np.arange(-10, 10, 0.01)
b=np.sin(x)
c=np.e**x
a=x**3
d=np.log(x)
plt.subplot(2, 2, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.plot(x, a, 'r--', label='x cubed')
plt.title('cubic')

plt.legend()
plt.subplot(2, 2, 2)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, b, 'g-.', label='sine of x')
plt.title('trigonometric')

plt.legend()
plt.subplot(2, 2, 3)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, c, 'y', label='e to the power of x')
plt.title('exponential')
plt.legend()
plt.subplot(2, 2, 4)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, d, label='log(x)')
plt.title('logarithmic')
plt.legend()
plt.subplots_adjust(left=0.1, right=0.9, 
                    top=0.9, bottom=0.1, 
                    wspace=0.4, hspace=0.4)
plt.show()
```

```python
x=np.random.randint(0, 10, 100)
y=np.random.randint(0, 10, 100)
plt.scatter(x, y, c='red', marker='x')
plt.title('random scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
arr=np.random.normal(size=1000)
plt.hist(arr, bins=30, alpha=0.7)
plt.title("Normal Histogram")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from matplotlib import colormaps as cm

fig=plt.figure()
ax=plt.axes(projection='3d')
x=np.linspace(-5, 5, 1000)
y=np.linspace(-5, 5, 1000)

xx, yy=np.meshgrid(x, y)
zz=np.cos(xx**2+yy**2)
surface=ax.plot_surface(xx, yy, zz, cmap='summer')
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('f(x, y)', fontsize=12)
plt.show()

```

```python
prods=['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales=[200, 150, 250, 175, 225]

plt.title('Sales stats')
plt.xlabel('The products')
plt.ylabel("The amount of sales")
colors = ['blue', 'green', 'red', 'blue', 'blue']
plt.bar(prods, sales, color=colors)
plt.show()
```

```python
import random
cats=np.array(['Category A', 'Category B', 'Category C'])
pers=np.array(['T1', 'T2','T3', 'T4'])
values1=np.array([random.randint(1, 10), random.randint(1, 10), random.randint(1, 10), random.randint(1, 10)])
values2=np.array([random.randint(1, 10), random.randint(1, 10), random.randint(1, 10), random.randint(1, 10)])
values3=np.array([random.randint(1, 10), random.randint(1, 10), random.randint(1, 10), random.randint(1, 10)])
plt.bar(pers, values1, label=cats[0], color='blue')
plt.bar(pers, values2, label=cats[1], bottom=values1, color='green')
plt.bar(pers, values3, label=cats[2], bottom=values1+values2, color='red')
plt.title('Random categories stacked')
plt.legend()
plt.xlabel('Time periods')
plt.ylabel('Amount sold')
```
